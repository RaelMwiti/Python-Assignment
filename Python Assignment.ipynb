18. Write a program that takes a list of numbers from the user (comma separated), converts them into a list of integers, and prints only the even numbers.


```python
##Ask user for comma-separated numbers
user_input = input("Enter numbers separated by commas: ")
 
##Convert input string into a list of integers
numbers = [int(num.strip()) for num in user_input.split(",")]
 
##Print only even numbers
print("Even numbers:")
for n in numbers:
    if n % 2 == 0:
        print(n)
```

28. Write a program that counts how many times each word appears in a list of words: Write a program that counts how many times each word appears in a list of words: ["apple", "banana", "apple", "cherry", "banana", "banana"].
.


```python
word = ["apple", "banana", "apple", "cherry", "banana", "banana"]
word_count = {}
for k in word:
    if k in word_count: 
        word_count[k] += 1
    else:
        word_count[k] = 1
for k, count in word_count.items():
    print(f"{k} : {count}")
```
